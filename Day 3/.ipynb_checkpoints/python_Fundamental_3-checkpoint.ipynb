# Conditional Statements

## 1. if else

```python
age = int(input("Enter Your age"))
if age>=18:
    print("You can Vote")
else:
    print("You Cannot Vote")
          
          
```

## 2. elif same as if else
### do not forget to add : 

```python
color = input("Enter the color : ")
if (color == "red"):
    print("Stop")
elif (color == "Yellow"):
    print("ready")
elif (color == "green"):
    print("go")
else:
    print("invalid color")
          
```

## Practicle Example of elif

```python
age = int(input("Enter Your Age"))
if(age<13):
    print("You are child")
elif (age>=13 and age<18):
    print("teenager")
else:
    print("adult")
```

## check whether a no. is multiple of 5  no or not

```python
num = int(input("Enter a number"))
if(num%5 == 0):
    print("yes")
else:
    print("no")
```

## 3. Match Case

```python
color = input("Enter the color")
match color:
    case "red":
        print("stop")
    case "green":
        print("go")
    case "yellow":
        print("ready")
    case _:
        print("Invalid Color")
        
        
```

## 4. While Loop

```python
count = 1
while (count<=5):
    print ("hello world", count)
    count+=1
print ("count is :",count)
```

## Multiplication of any no Enter by user

```python
num = int(input("Enter any no."))
i = 1;
while(i<=10):
    print (num," x ",i,"=",num*i)
    i+=1
```

## 5. continue and break

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

## Count the no of 'i' in artificial intelligence

```python
str = "artificial intelligence"
count = 0
for var in str:
    if(var=='i'):
        count+=1;

print(count)
```

## 8. range(start, stop, step)
### start → starting number (default is 0)
### stop → ending number (not included)
### step → difference between numbers (default is 1)

```python
for i in range(2, 6):
    print(i)
```

```python
for i in range(1,8,2):
    print(i)
```

## Sum of n Natural Number

```python
n = int(input("Enter a +ve integer"))
sum = 0;
for i in range(1, n+1):
    sum+=i
print("sum", sum)
```

## 9. function

```python
def add(a,b):
    return a+b

print(add(4,5))
```

## 10. Default Parameter

```python
def add(a,b=9):
    return a+b

print(add(4))
```

## 11. Lambda Function

```python
sum = lambda a,b,c: a+b+c
print(sum(2,3,5))
```

## 12. Factorial of any no. n

```python

def fac(n):
    if(n==1 or n==0):
        return 1
    return n*fac(n-1)
print(fac(5))
```

```python
fact = 1
for i in range(1, 5+1):
    fact*=i
print(fact)
```

```python
ans = 1
i = 1
while(i<=5):
    ans = ans*i
    i+=1
print(ans)
```
