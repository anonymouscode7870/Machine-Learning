## Dictionary in python
- values are mutable
- order doesn't matter
- its like javascript obj

```python
info = {
    "name":"abc",
    "age": 6,
    "subject":["maths", "science"],
    88: "hello ji"
}
print(info)
print(info["name"])
info["age"] = 8
print(info["age"])
```

## Methods in dictionary

```python
# get all keys
key = info.keys()
print(key)
print(type(key))

key1 = list(info.keys())
print(key1)
print(type(key1))
```

```python
# get all values
val = info.values()
print(val)
print(type(val))

val1 = list(info.values())
print(val1)
print(type(val1))
```

```python
# get key values pairs
print(info.items())
```

```python
# get method
print(info.get("name")) # another way to access values if key does not exist then it prints none and program do not terminates
```

```python
# update method add new item
info["fds"] = "dw"
# or
info.update(
    {
        "new":"old"
    }
)
```

```python
print(info)
```

## Sets in python
- No duplicate elements
- Unordered (no indexing)
- Mutable (can add/remove elements)
- Elements must be immutable (numbers, strings, tuples)

```python
s = {1 ,2, 3, 3, 3,2,5}
print(s)
print(len(s))
```

```python
s.add(9)
print(s)
```

```python
# empty dic
s1 = {};
type(s1)
```

```python
# empty sets
s2 = set()
print(type(s2))
```

## Methods in Sets

- `s.add(val)`  
  Adds `val` to the set

- `s.remove(val)`  
  - Removes `val` from the set  
  - Raises an error if `val` is not present

- `s.clear()`  
  - Removes all elements from the set (empties it)

- `s.pop()`  
  - Removes and returns a random element from the set  
  - Raises an error if the set is empty

- `s.union(set2)`  
  - Returns a new set containing all elements from both sets  
  - Original sets remain unchanged

- `s.intersection(set2)`  
  - Returns a new set containing only common elements  
  - Original sets remain unchanged


```python
s = {1, 2, 3}
t = {3, 4, 5}

print(s.union(t))        # {1, 2, 3, 4, 5}
print(s.intersection(t)) # {3}
```
