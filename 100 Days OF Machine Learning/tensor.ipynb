# Tensor
### A tensor is a mathematical object used to store and represent data in multiple dimensions.
### A tensor is a multidimensional array of numbers.


## Scalar or 0-D Tensor

```python
import numpy as np
a = np.array(3)
a
```

```python
a.ndim
```

## 1D Tensor/Vector/array  (no. of axis = Rank = no. of dimension)

```python
a = np.array([1,2,3])
```

```python
a
```

```python
a.ndim
```

## 2D Tensor/Matrices

### 1D tensor/vector is the collection of scalar
### 2D tensor is the collection of vector or 1d tensor

### collection of 2D tensor is 3d tensor
### collection of 3D tenor is 4d tensor

```python
mat = np.array([[1,2,3], [4,5,6]])
mat
```

```python
mat.ndim
```

### Shape means no. of Items you can store in row, cols i.e., shape(m,n)
### size = m*n

```python
tensor3d = np.array([
    [
        [1, 2, 3],
        [4, 5, 6],
        [1, 2, 3],
        [4, 5, 6]
    ],
    [
        [7, 8, 9],
        [1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]
    ]
])

tensor3d.shape
```
